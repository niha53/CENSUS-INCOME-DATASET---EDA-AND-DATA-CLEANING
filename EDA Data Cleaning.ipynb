# EDA-6: Identify the Ouliers

```python
# Def:Outliers are extreme values that deviates significantly from rest of the data.

## Interview Question: IS categorical column can have outlier?

# 1. IQR(Interquartile Range)
# 2. Boxplot
```

```python
# IQR(Interquartile Range): 50% of data

IQR = Q3 - Q1

Q1: (First Quantile, 25th)
Q3: (Third Quantile, 75th)

lower_bound(Min) = Q1 - 1.5 * IQR
upper_bound(Max) = Q3 + 1.5 * IQR
```

```python
Q1 = df.age.quantile(.25)
```

```python
Q3 = df.age.quantile(.75)
```

```python
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_bound = Q1 -1.5 * IQR
upper_bound = Q3 +1.5 * IQR
```

```python
lower_bound,upper_bound
```
